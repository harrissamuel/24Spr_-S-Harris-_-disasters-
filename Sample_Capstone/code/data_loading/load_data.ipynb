```python
## Slice data so it only includes dates from 2000 onward and get rid of some columns
_d = df[df['Start Year'] >= 2000]
print(_d.shape)
g = ['DisNo.', 'Classification Key', 'Disaster Group',
    'Disaster Subgroup', 'Disaster Type', 'Disaster Subtype', 'Event Name', 'ISO', 'Country', 'Subregion', 'Region',
    'Location', 'Magnitude','Magnitude Scale', 'Start Year', 'Start Month', "Total Damage, Adjusted ('000 US$)", 'Total Deaths' ]
_d = _d[g]
```

```python
# check NAs for filtered data
nan_count = _d.isna().sum()
print(nan_count)
```

```python
# check columns
print(_d.columns)
_d.shape
```

```python
# value count for disaster type
_d['Disaster Type'].value_counts()
```

```python
# NAs in the flood disaster type
_d[_d['Disaster Type']== 'Flood'].isna().sum()
```

```python
# Lack of data showing zero deaths. Look into metadata about this
_d[_d['Total Deaths']== 0].count()
```

```python
# NAs for deaths by disaster type
print(_d[_d['Total Deaths'].isna()].groupby('Disaster Type').size())
```

```python
# World Bank Data
# https://databank.worldbank.org/indicator/NY.GDP.PCAP.CD/1ff4a498/Popular-Indicators#
gdp = pd.read_csv('gdp_per_cap.csv', skiprows=[218, 219, 220, 221, 222])
gdp.head()
gdp.tail()
```

```python
# How many disaster events per country?
_d['ISO'].value_counts()
```

```python
# Wrangle the gdp data to make it longer
gdp = pd.melt(gdp, id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], var_name='Year', value_name='gdp_per_cap')
```

```python
## Clean up year column
gdp.head()
gdp['Year'] = gdp['Year'].str[:4].astype(int)
```

```python
_d['Start Year'].head()
gdp.head()
```

```python
## Merge the World Bank Data with EM Dat data
merged_df = pd.merge(_d, gdp[['Year', 'Country Code', 'gdp_per_cap']], left_on=['Start Year', 'ISO'], right_on=['Year', 'Country Code'])
merged_df = merged_df.drop(['Year', 'Country Code'], axis=1)
merged_df
```

```python
# Check Values in em-Data and not in gdp data
values_not_in_gdp = _d['ISO'][~_d['ISO'].isin(gdp['Country Code'])]

print("Categorical values in _d not in gdp:")
print(values_not_in_gdp.value_counts())
```

```python
# Check Values in gdp but not in _d (EM Dat)
values_not_in_d = gdp['Country Code'][~gdp['Country Code'].isin(_d['ISO'])]
print("Categorical values in gdp not in _d]:")
print(values_not_in_d.value_counts())
## Taiwan and other countries filtered out
```

```python
## Dataframe with GDP per cap info along with disasters
df = merged_df
df
```
