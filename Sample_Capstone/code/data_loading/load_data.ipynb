```python
import matplotlib.pyplot as plt
import numpy as np

def boxplot(column):
    # Drop NaNs from the column
    column = column.dropna()
    
    # Calculate mean, median, and standard deviation
    mean = column.mean()
    median = column.median()
    std_dev = column.std()
    percentile_25 = np.percentile(column, 25)
    percentile_75 = np.percentile(column, 75)

    # Create the boxplot without outliers
    plt.figure(figsize=(8, 6))
    plt.boxplot(column, vert=False, showfliers=False)
    plt.title('Boxplot with Mean, Median, and Percentiles (Excluding Outliers)')
    plt.xlabel('Values')
    plt.yticks([])
    plt.grid(True)

    # Add mean, median, and standard deviation to the plot
    plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
    plt.axvline(median, color='green', linestyle='-.', label=f'Median: {median:.2f}')
    plt.axvline(mean + std_dev, color='blue', linestyle=':', label=f'Standard Deviation: {std_dev:.2f}')
    plt.axvline(mean - std_dev, color='blue', linestyle=':')
    plt.axvline(percentile_25, color='purple', linestyle='-', label=f'25th Percentile: {percentile_25:.2f}')
    plt.axvline(percentile_75, color='orange', linestyle='-', label=f'75th Percentile: {percentile_75:.2f}')
    plt.legend()

    plt.show()



import seaborn as sns
import matplotlib.pyplot as plt

def scatterplot(x, y, title, x_lim=None, y_lim=None):
    # Create scatterplot using Seaborn
    sns.scatterplot(x=x, y=y)
    
    # Add labels and title
    plt.xlabel(x.name)
    plt.ylabel(y.name)
    plt.title(title)
    
    # Set limits if provided
    if x_lim:
        plt.xlim(x_lim)
    if y_lim:
        plt.ylim(y_lim)
    
    # Show plot
    plt.show()
```

```python
def filter_df(df, series, quantile_low, quantile_high):
    # Calculate the quantile threshold
    threshold_1 = series.quantile(quantile_low)
    threshold_2 = series.quantile(quantile_high)
    
    # Filter the series to exclude outliers and retain values above the threshold
    filtered_df = df[(series > threshold_1) & (series < threshold_2)]
    
    return filtered_df

df_filtered= filter_df(df, df['Total Deaths'], .25, .99)

boxplot(df_filtered['Total Deaths'])
boxplot(df_filtered["Total Damage, Adjusted ('000 US$)"])
```

```python
from scipy.stats import percentileofscore
import numpy as np

def scale_mapper(data):
    # Convert the column to numeric type
    data.iloc[:, 5] = pd.to_numeric(data.iloc[:, 5], errors='coerce')
    
    # Calculate percentiles for non-NaN values
    valid_values = data.iloc[:, 5].dropna()
    percentiles = percentileofscore(valid_values, valid_values)
    
    # Map the percentile rank to a 1 to 10 scale
    mapped_values = ((percentiles / 10).astype(int)) + 1
    
    # Create a new column with mapped values
    data['mapped_values'] = np.nan
    data.loc[data.iloc[:, 5].notna(), 'mapped_values'] = mapped_values
    
    # Replace mapped values of 11 with 10
    data.loc[data['mapped_values'] == 11, 'mapped_values'] = 10
    
    return data
```

```python
scale_mapper(gdp)
scale_mapper(i_mort)
scale_mapper(electy)
scale_mapper(pop_den)
scale_mapper(co2)
scale_mapper(lf_expect)
```
