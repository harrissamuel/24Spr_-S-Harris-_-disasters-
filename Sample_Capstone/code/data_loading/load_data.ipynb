```python
## Slice data so it only includes dates from 2000 onward and get rid of some columns
_d = df[df['Start Year'] >= 2000]
print(_d.shape)
g = ['DisNo.', 'Classification Key', 'Disaster Group',
    'Disaster Subgroup', 'Disaster Type', 'Disaster Subtype', 'Event Name', 'ISO', 'Country', 'Subregion', 'Region',
    'Location', 'Magnitude','Magnitude Scale', 'Start Year', 'Start Month', "Total Damage, Adjusted ('000 US$)", 'Total Deaths' ]
_d = _d[g]
```

```python
# check NAs for filtered data
nan_count = _d.isna().sum()
print(nan_count)
```

```python
# check columns
print(_d.columns)
_d.shape
```

```python
# value count for disaster type
_d['Disaster Type'].value_counts()
```

```python
# NAs in the flood disaster type
_d[_d['Disaster Type']== 'Flood'].isna().sum()
```

```python
# Lack of data showing zero deaths. Look into metadata about this
_d[_d['Total Deaths']== 0].count()
```

```python
# NAs for deaths by disaster type
print(_d[_d['Total Deaths'].isna()].groupby('Disaster Type').size())
```
