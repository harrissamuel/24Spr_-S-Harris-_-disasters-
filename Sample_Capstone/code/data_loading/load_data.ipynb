```python
# Prepare Data for mergings
pop_den = pd.read_csv('pop_den.csv')
i_mort = pd.read_csv('infant_mort.csv')
# pop den
pop_den = pd.melt(pop_den, id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], var_name='Year', value_name='pop_den')
pop_den['Year'] = pop_den['Year'].str[:4].astype(int)
# i_mort
i_mort = pd.melt(i_mort, id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], var_name='Year', value_name='i_mort')
i_mort['Year'] = i_mort['Year'].str[:4].astype(int)
i_mort.head()
```

```python
merged_df = pd.merge(df, pop_den[['Year', 'Country Code', 'pop_den']], left_on=['Start Year', 'ISO'], right_on=['Year', 'Country Code'])
df = merged_df.drop(['Year', 'Country Code'], axis=1)

merged_df = pd.merge(df, i_mort[['Year', 'Country Code', 'i_mort']], left_on=['Start Year', 'ISO'], right_on=['Year', 'Country Code'])
df = merged_df.drop(['Year', 'Country Code'], axis=1)
```

```python
df['gdp_per_cap'].value_counts()
```

```python
# Make missing values NAs
df['gdp_per_cap'] = df['gdp_per_cap'].apply(pd.to_numeric, errors='coerce')
df['pop_den'] = df['pop_den'].apply(pd.to_numeric, errors='coerce')
df['i_mort'] = df['i_mort'].apply(pd.to_numeric, errors='coerce')


print(df['gdp_per_cap'].isna().sum())
print(df['pop_den'].isna().sum())
print(df['i_mort'].isna().sum())
```

```python
# Prepare Data for mergings
electy = pd.read_csv('acess_elec_percent.csv')
co2 = pd.read_csv('co2_emission_metric_ton_per_cap.csv')
lf_expect = pd.read_csv('life_expectancy.csv')
# get only countries
electy = electy.loc[:216,:]
co2 = co2.loc[:216,:]
lf_expect = lf_expect.loc[:216,:]
```

```python
# co2
co2 = pd.melt(co2, id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], var_name='Year', value_name='co2_emissions')
co2['Year'] = co2['Year'].str[:4].astype(int)

# electricity
electy = pd.melt(electy, id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], var_name='Year', value_name='electricity_percentage')
electy['Year'] = electy['Year'].str[:4].astype(int)

# life 
lf_expect = pd.melt(lf_expect, id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], var_name='Year', value_name='life_expectancy')
lf_expect['Year'] = lf_expect['Year'].str[:4].astype(int)
```

```python
# co2
merged_df = pd.merge(df, co2[['Year', 'Country Code', 'co2_emissions']], left_on=['Start Year', 'ISO'], right_on=['Year', 'Country Code'])
df = merged_df.drop(['Year', 'Country Code'], axis=1)

# electricity
merged_df = pd.merge(df, electy[['Year', 'Country Code', 'electricity_percentage']], left_on=['Start Year', 'ISO'], right_on=['Year', 'Country Code'])
df = merged_df.drop(['Year', 'Country Code'], axis=1)

# life 
merged_df = pd.merge(df, lf_expect[['Year', 'Country Code', 'life_expectancy']], left_on=['Start Year', 'ISO'], right_on=['Year', 'Country Code'])
df = merged_df.drop(['Year', 'Country Code'], axis=1)
```

```python
df.head()
```

```python
# Make missing values NAs
df['co2_emissions'] = df['co2_emissions'].apply(pd.to_numeric, errors='coerce')
df['electricity_percentage'] = df['electricity_percentage'].apply(pd.to_numeric, errors='coerce')
df['life_expectancy'] = df['life_expectancy'].apply(pd.to_numeric, errors='coerce')


print(df['co2_emissions'].isna().sum())
print(df['electricity_percentage'].isna().sum())
print(df['life_expectancy'].isna().sum())
```
